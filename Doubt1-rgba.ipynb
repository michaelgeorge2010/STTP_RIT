```python
img = cv2.imread('rgba.png',cv2.IMREAD_UNCHANGED)
img.shape
```

```python
a = img[:,:,3]
bluec = img[:,:,2]
greenc = img[:,:,1]
redc = img[:,:,0]
```

```python
from matplotlib.pyplot import *
```

```python
figure(figsize=(18,4))
subplot(1,5,1)
imshow(img),title('original')

subplot(1,5,2)
imshow(a,cmap='gray'),colorbar(),title('a channel')

subplot(1,5,3)
imshow(bluec,cmap='gray'),colorbar(),title('blue channel')

subplot(1,5,4)
imshow(greenc,cmap='gray'),colorbar(),title('green channel')

subplot(1,5,5)
imshow(redc,cmap='gray'),colorbar(),title('red channel')
```

```python
img2 = cv2.imread('butterfly.jpg',cv2.IMREAD_UNCHANGED)
```

```python
redc2 = img2[:,:,0]
greenc2 = img2[:,:,1]
bluec2 = img2[:,:,2]
```

```python
figure(figsize=(16,4))
subplot(1,4,1)
imshow(img2),title('original')

subplot(1,4,2)
imshow(redc2,cmap='gray'),colorbar(),title('red channel')

subplot(1,4,3)
imshow(greenc2,cmap='gray'),colorbar(),title('green channel')

subplot(1,4,4)
imshow(bluec2,cmap='gray'),colorbar(),title('blue channel')
```
