```python
img = cv2.imread('eiffel.jpg')
rows,cols,chans = img.shape

startpts =  float32([[100, 100], [330,100],  [330,400], [100,400]])
endpts   =  float32([[362, 107], [530,139] , [453,416], [319,337]])
M = cv2.getPerspectiveTransform(startpts,endpts)

dst = cv2.warpPerspective(img,M,(rows,cols))

figure(figsize = (16,8))
subplot(1,2,1)
imshow(img)
subplot(1,2,2)
imshow(dst)
```
